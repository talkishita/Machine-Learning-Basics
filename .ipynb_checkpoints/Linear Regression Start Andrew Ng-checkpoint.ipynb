```python
w=200
b=100
print(f"w : {w}")
print(f"b : {b}")
```

now we need to make a straight line which will give us y cap.
To do that we need a function 
where we have 3 input from the equation  f(x)=wx+b to curate the answer that is w,x and b.
Now we have x as value from size array that is index i from x axis (size) to curate this ans
we need w and b as well whcih are basically take magnitudes here (scalar)= represents magnitude


```python
def model_output(x,w,b):
    """
    parameters are-> x,w,b
    when we provide them value while we call them then they are arguments
    Args:-
        x(ndarray (m, )) :Numpy n dimensional array with m shape
        w,b(scalar) : model parameters :magnitude
    Returns:-
        y hat
        y(ndarray(m,)) : target values with each different value of x
    """
    m=x.shape[0] # size of the array
    f_d=np.zeros(m)
    # np is nupmy, zeroes is the initialization of an array of size m with 0's as the inputs if m=2 then {0,0}
    #in python i doesn't need to be defined when we initialize i
    for i in range(m):
        f_d[i]=w*x[i]+b
    return f_d
```

```python
temp_f=model_output(x_train,w,b)
plt.plot(x_train,temp_f, c='b', label='Our Predictions')
# plot is used to plot a line
plt.scatter(x_train,y_train, marker="x",c="r",label='Actual Value')
# sactter points on the graph
plt.title("Housing Predition")
plt.xlabel("Size")
plt.ylabel("Prices")
plt.legend(loc=2)
# legend's default is to place the label in uppe right and making it 2 will move it to upper left
plt.show()
         
```
